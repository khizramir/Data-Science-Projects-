```python
stem_cats = ['Biology','Computer Science', 'Engineering',
             'Math and Statistics', 'Physical Sciences','Psychology']

lib_arts_cats = ['Art and Performance','Communications and Journalism',
             'English','Foreign Languages','Social Sciences and History']

other_cats = ['Agriculture', 'Architecture','Education',
              'English','Health Professions','Public Administration']
```

```python
fig = plt.figure(figsize = (18,25))

# in matplotlib when you make plots, it runs horizontally
# therefore subplot (6,3,1), (6,3,4), (6,4,7) coz we are
# focusing on just one category (stem_cats)

for i in range(0,16,3):
    #print(i+1)
    ind = int(i/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees["Year"],women_degrees[stem_cats[ind]], label = "Women")
    ax.plot(women_degrees["Year"],100-women_degrees[stem_cats[ind]], label = "Men")
    ax.set_title(stem_cats[ind])
    ax.set_ylim(0,100)
    ax.set_xlim(1970,2012)
    ax.set_yticks([0,100])
    ax.tick_params(bottom = False, labelbottom = False)
    for value in ax.spines.values():
        value.set_visible(False)
    if i == 15: 
          ax.tick_params(labelbottom = True)
    ax.legend()
    
for j in range(1,16,3):
    #print(j+1)
    index = int(j/3)
    ax = fig.add_subplot(6,3,j+1)
    ax.plot(women_degrees["Year"],women_degrees[lib_arts_cats[index]], label = "Women")
    ax.plot(women_degrees["Year"],100-women_degrees[lib_arts_cats[index]], label = "Men")
    ax.set_title(stem_cats[index])
    ax.set_ylim(0,100)
    ax.set_xlim(1970,2012)
    ax.set_yticks([0,100])
    ax.tick_params(bottom = False, labelbottom = False)
    for value in ax.spines.values():
        value.set_visible(False)
    if index == 4: 
          ax.tick_params(labelbottom = True)
    ax.legend()   
    
for x in range(2,18,3):
    #print(x+1)
    indx = int(x/3)
    ax = fig.add_subplot(6,3,x+1)
    ax.plot(women_degrees["Year"],women_degrees[other_cats[indx]], label = "Women")
    ax.plot(women_degrees["Year"],100-women_degrees[other_cats[indx]], label = "Men")
    ax.set_title(other_cats[indx])
    ax.set_ylim(0,100)
    ax.set_xlim(1970,2012)
    ax.set_yticks([0,100])
    ax.tick_params(bottom = False, labelbottom = False)
    for value in ax.spines.values():
        value.set_visible(False)
    if indx == 5: 
         ax.tick_params(labelbottom = True)
    ax.legend()

plt.show()


    
```
